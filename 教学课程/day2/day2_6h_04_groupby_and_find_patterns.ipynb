# Day 2 模块 4：分组强化与找规律

昨天我们已经见过 `groupby()`。今天继续把它用得更熟。


```python
from pathlib import Path
import pandas as pd

candidate_paths = [
    Path('day2_cafe_sales.csv'),
    Path('day2') / 'day2_cafe_sales.csv',
    Path('教学课程') / 'day2' / 'day2_cafe_sales.csv',
]
for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day2_cafe_sales.csv')

df = pd.read_csv(csv_path)
print(csv_path.resolve())
```

## 2. 不同天气的平均营收

先回答：不同天气下，平均营收一样吗？


```python
weather_order = ['晴', '多云', '阴', '小雨', '大雨']
df.groupby('weather_label')['sales'].mean().reindex(weather_order).round(2)
```

## 3. 周末和工作日

再回答：周末真的卖得更好吗？


```python
df.groupby('is_weekend')['sales'].mean().round(2)
```

## 4. 促销和不促销

促销到底值不值？


```python
df.groupby('promotion')['sales'].mean().round(2)
```
