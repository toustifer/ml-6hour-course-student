# Day 2 模块 2：选列、看行、轻量 loc/iloc

现在开始真正动手操作 DataFrame。


```python
from pathlib import Path
import pandas as pd

candidate_paths = [
    Path('day2_cafe_sales.csv'),
    Path('day2') / 'day2_cafe_sales.csv',
    Path('教学课程') / 'day2' / 'day2_cafe_sales.csv',
]
for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day2_cafe_sales.csv')

df = pd.read_csv(csv_path)
print(csv_path.resolve())
```

## 1. 只看一列

如果我们现在只关心营收，可以只拿 `sales` 这一列。


```python
df['sales'].head()
```

## 2. 同时看两列

如果想同时看天数和营收，可以拿多列。


```python
df[['day', 'sales']].head()
```

## 3. 看前面和后面

`head()` 看前面，`tail()` 看后面。


```python
df[['day', 'sales']].tail()
```

## 4. 按位置取几行：`iloc`

`iloc` 是按位置编号取。


```python
df.iloc[0:5]
```

## 5. 按标签看列：`loc`

`loc` 更适合按列名来取。


```python
df.loc[0:4, ['day', 'weather_label', 'sales']]
```
