# Day 2 模块 3：条件过滤与排序

现在开始回答“我只想看某一部分数据”的问题。


```python
from pathlib import Path
import pandas as pd

candidate_paths = [
    Path('day2_cafe_sales.csv'),
    Path('day2') / 'day2_cafe_sales.csv',
    Path('教学课程') / 'day2' / 'day2_cafe_sales.csv',
]
for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day2_cafe_sales.csv')

df = pd.read_csv(csv_path)
print(csv_path.resolve())
```

## 1. 只看高营收天

如果只想看营收大于等于 3000 的天，可以用条件过滤。


```python
df[df['sales'] >= 3000].head()
```

## 2. 只看晴天

文字列也可以过滤。


```python
df[df['weather_label'] == '晴'].head()
```

## 3. 同时加两个条件

例如：只看周末并且营收较高的天。


```python
df[(df['is_weekend'] == 1) & (df['sales'] >= 3000)].head()
```

## 4. 排序：谁卖得最好

`sort_values()` 可以帮助我们找出最高或最低。


```python
df.sort_values('sales', ascending=False)[['day', 'weather_label', 'sales']].head(10)
```
