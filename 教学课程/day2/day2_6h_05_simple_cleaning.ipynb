## 1. 什么叫缺失值

缺失值就是某个位置本来应该有数据，但现在没有。

先自己造一个小例子来理解。


```python
sample = pd.DataFrame({'sales': [3200, None, 2800, 3500], 'price': [28.5, 30.0, None, 26.0]})
sample
```

## 2. 用 `fillna()` 简单补上

这里只做最简单的演示：空的位置先补成一个固定值。


```python
sample.fillna(0)
```

## 3. 什么叫异常值

异常值就是特别离谱、明显和其它数据差很多的值。


```python
sample2 = pd.DataFrame({'sales': [3200, 3300, 3100, 99999, 3000]})
sample2
```

## 4. 用条件先找出来

这里只做直观处理：先把特别大的值找出来。


```python
sample2[sample2['sales'] > 10000]
```
