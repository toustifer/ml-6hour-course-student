## 1. 一句话理解机器学习

以前常见的做法是：人自己写规则。

```text
如果晴天 且 周末 且 有促销 → 营收可能比较高
```

机器学习更像是：

> 把很多天的历史数据交给模型，让模型自己总结规律，再用来预测新的一天。

今天我们要做的，就是用咖啡馆历史数据，训练一个能预测 `sales` 的模型。


## 1.5 和前面那条最简单模型对上号

前面我们已经接触过一种最简单的模型感觉：

```text
Y = KX + B
```

可以先这样记：

- **X**：输入信息（比如某个影响营收的因素）
- **Y**：输出结果（比如营收）
- **K、B**：模型要学的参数（斜率、截距）
- 训练时会看预测和真实差多少（**损失**变小）
- 训完后还要看它在新数据上好不好（**评估**）

更复杂的模型，也可以先理解成：

> 把大问题拆成很多小计算，再用函数一步步拼起来。

有的课会把“一个小计算单元”叫作神经元。今天**不要求**你会推神经元公式，只要知道：

- 直线模型：用一条直线去拟合
- 决策树：用一串 if/else 去切分
- 随机森林：很多棵树一起平均，通常更稳

同一件事的不同做法，目标都是：**从数据里学规律，预测新的 Y**。


## 2. 监督学习（今天主线）与对照

我们有“题目”和“答案”：

- 题目：天气、价格、是否周末、是否促销……（特征）
- 答案：当天营收 `sales`（目标）

这种“有答案可对照学习”的方式，叫**监督学习**。

今天做的是监督学习里的**回归**：预测一个连续数字（营收多少钱）。

再对照记两个词（今天不深挖）：

| 说法 | 白话 | 今天？ |
| --- | --- | --- |
| **监督学习** | 有标准答案可对照 | ✅ 主线 |
| **无监督学习** | 没有标准答案，找结构（如把相似的天聚成几堆） | 了解名字即可 |
| **回归** | 预测连续数字（营收） | ✅ 主线 |
| **分类** | 预测类别（如“高/低营收档”“会不会排队”） | 知道区别即可 |

一句话：

> 今天：用有答案的历史数据，学一个**回归**模型猜营收。


## 3. 特征 X 和目标 y

- **特征 X**：模型用来判断的信息列
- **目标 y**：模型要预测的结果列

在这门课里：

- `X`：价格、促销、周末、气温、天气评分……
- `y`：`sales` 营收


## 4. 过拟合与欠拟合

### 过拟合：自己考自己考满分

如果模型只把训练集背下来，换一批新数据就不会了，这就叫**过拟合**。

白话：

> 模拟考试全会，正式考试换题就不会。

### 欠拟合：题太难 / 模型太简单

反过来，模型连训练集都学得很差，规律几乎没抓住，叫**欠拟合**。

白话：

> 书都没看懂，考试当然不行。

| | 训练集 | 测试集 | 感觉 |
| --- | --- | --- | --- |
| 欠拟合 | 差 | 差 | 太简单，没学会 |
| 较合适 | 还行 | 还行 | 能泛化 |
| 过拟合 | 很好 | 明显变差 | 背题了 |

所以我们要把数据分成：

- **训练集**：让模型学习
- **测试集**：检查模型是不是真的学会了

今天用浅树、`max_depth`、随机森林，都和“别背太死 / 别太弱”有关。


## 5. 今天的路线图

```text

```
