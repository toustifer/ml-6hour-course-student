# Day 5 模块 5：给老板的汇报话术（边点边讲）

目标：3～5 分钟，**有数字、有点击、有建议**，不是念代码。

## 1. 推荐时间轴

| 时间 | 你说 / 做 | 屏幕 |
| --- | --- | --- |
| 0:00–0:30 | 我回答哪 2～3 个经营问题 | 笔记本或一页纸 |
| 0:30–1:30 | 1～2 条场景发现（带数字） | notebook 图/表 或 口头 |
| 1:30–2:00 | 模型白话：R²/MAE + Top 因素 | 网页「训练」结果 |
| 2:00–3:30 | 现场 what-if：只拧 1～2 个旋钮 | 网页「预测」 |
| 3:30–4:30 | 3 条建议 + 如何验证 | 口头 |
| 4:30–5:00 | 局限一句 | 口头 |

## 2. 开场句（可抄改）

```text
我用大约 200 天的经营记录，做了场景对比和营收预测模型。
今天想汇报三件事：我们发现了什么、模型更盯什么、
以及如果改促销/竞品，预测会怎么动——我打开演示给你看。
```

## 3. 演示时的规矩

1. **先讲问题，再点按钮**（别一上来狂点）
2. 一次只改**一个**输入，让老板看清因果感（仍是推演，不是保证）
3. 预测数字报整数即可，强调「比较高低」
4. 主动说局限：「模拟数据 / 有误差 / 不是严格因果」

## 5. 反面教材（避免）

- 全程盯代码滚动，老板看不见结论
- 只说「我用了机器学习」没有任何数字
- 一次改五个框，说不清是谁导致变化
- 把预测说成「明天一定卖这么多」
