## 1. 数据流（背这张）

```text
cafe_sales.csv
  → load_and_clean()     读表、列名统一、天气分、简单清洗
  → train_model()        随机森林 fit，算 R² / 重要性
  → MODEL_CACHE          记在内存里
  → /api/*               把结果变成 JSON
  → index.html           表格、图表、表单
```

## 2. 主要 API（完整版）

| 路径 | 方法 | 作用 |
| --- | --- | --- |
| `/` | GET | 返回网页 |
| `/api/data/summary` | GET | 总天数、平均营收等摘要 |
| `/api/data/table` | GET | 分页表格数据 |
| `/api/data/overview` | GET | 每日营收序列（画趋势） |
| `/api/train` | POST | 训练，返回 R²、MSE、重要性 |
| `/api/predict` | POST | 输入特征，返回预测营收 |
