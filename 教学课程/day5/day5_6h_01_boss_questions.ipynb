## 1. 老板问题墙

| 类型 | 老板可能问 | 你用什么应答 |
| --- | --- | --- |
| 整体 | 一天大概卖多少？波动大吗？ | 均值 / 最好最差天 |
| 场景 | 下雨、周末、促销差在哪？ | `groupby` / 交叉表 |
| 结构 | 什么最影响营收？ | 特征重要性（业务话） |
| 推演 | 竞品多了 / 做促销会怎样？ | what-if（notebook 或网页） |
| 行动 | 下周先改什么？ | 3 条可验证建议 |
| 可信 | 模型靠谱吗？ | R² / MAE 用白话说 |

## 2. 汇报禁用 / 推荐

| 少说 | 多说 |
| --- | --- |
| 「我跑了随机森林」 | 「测试集上大约能解释七成多波动」 |
| 「R² 是 0.76」单独一句 | 「预测常偏几百块很正常，用来排优先级」 |
| 只贴一张巨大表格 | 1 个结论 + 1 个数字 + 1 个动作 |

## 3. 好建议三句话

```text
1. 发现：数据里看到什么（带数字）
2. 解释：可能因为什么（谨慎，别硬因果）
3. 行动：建议试什么、怎么验证
```

## 5. 准备数据


````python
from pathlib import Path
import pandas as pd

candidate_paths = [
    Path('day5_cafe_sales.csv'),
    Path('day5') / 'day5_cafe_sales.csv',
    Path('教学课程') / 'day5' / 'day5_cafe_sales.csv',
]
for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day5_cafe_sales.csv')

df = pd.read_csv(csv_path)
print(csv_path.resolve())
print('shape:', df.shape)
df.head()
````

````python
weather_score_map = {'晴': 1.0, '多云': 0.8, '阴': 0.6, '小雨': 0.4, '大雨': 0.3}
df = df.copy()
df['weather_score'] = df['weather_label'].map(weather_score_map)
df['weekend_label'] = df['is_weekend'].map({0: '工作日', 1: '周末'})
df['promo_label'] = df['promotion'].map({0: '无促销', 1: '有促销'})
print(df[['day', 'weather_label', 'weekend_label', 'promo_label', 'sales']].head())
print('平均营收:', round(df['sales'].mean(), 1))
````
