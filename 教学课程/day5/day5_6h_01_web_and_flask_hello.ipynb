## 1. 三个角色（先记人话）

| 角色 | 好比 | 今天对应 |
| --- | --- | --- |
| **浏览器（客户端）** | 你点餐的窗口 | Chrome / Edge |
| **服务器** | 后厨 | 你电脑上跑的 `python app.py` |
| **接口 / 路由** | 菜单上的菜名 | `@app.route('/api/train')` |

请求怎么走：

```text
你在浏览器输入地址或点按钮
  → 服务器找到对应函数
  → 函数算完（可能训模型、查表）
  → 把结果（HTML 页面 或 JSON 数据）送回浏览器
```

### GET 和 POST（够用版）

| | 常见用途 | 今天 |
| --- | --- | --- |
| **GET** | 打开页面、取数据 | 打开首页、`/api/hello` |
| **POST** | 提交表单、触发训练 | `/api/train`、`/api/predict` |

### JSON

前后端传结构化数据常用 JSON，长得像 Python 字典：

```text
{"status": "success", "r2": 0.76}
```


## 2. Flask 是什么

Flask = 用 Python 写网站的**轻量框架**。

最小骨架：

```python
from flask import Flask
app = Flask(__name__)

@app.route('/')
def home():
    return '你好'

app.run(port=5000)
```

- `@app.route('/')`：访问网站根路径时，调用下面的函数
- `return` 的字符串会显示在浏览器里
