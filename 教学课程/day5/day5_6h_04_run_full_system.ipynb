## 5. 常见故障

| 现象 | 处理 |
| --- | --- |
| `No module named flask` | `conda install flask -y`，内核/终端都用 `day1_ml` |
| 地址无法访问 | 是否在 `system` 目录启动？终端是否还在跑？ |
| 预测失败 | 是否先点了训练？ |
| 端口占用 | 关掉另一个 Flask；或改 `port=` |
| 中文乱码 | 浏览器编码 UTF-8；CSV 用 utf-8 |
