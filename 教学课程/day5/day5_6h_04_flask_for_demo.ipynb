## 3. 浏览器访问时发生了什么（极简）

```text
打开 http://127.0.0.1:5001
  → Flask 返回一个页面（表单）
点「训练」
  → 浏览器请求 /api/train → Python 训 RF → 返回 R² / 重要性
点「预测」
  → 浏览器请求 /api/predict → Python predict → 返回营收猜测
```

| 词 | 白话 |
| --- | --- |
| 端口 5001 | 本机上这个小网站的门牌号 |
| API | 网页和 Python 约定好的「要数据」入口 |
| 本地 | 只在你电脑上跑，不是上公网 |
