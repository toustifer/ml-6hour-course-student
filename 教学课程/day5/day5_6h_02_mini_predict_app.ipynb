## 5. 用 notebook 调迷你 API（服务需已启动）


````python
import json
import urllib.request

def post_json(url, payload=None):
    data = None if payload is None else json.dumps(payload).encode('utf-8')
    req = urllib.request.Request(
        url, data=data, method='POST',
        headers={'Content-Type': 'application/json'},
    )
    with urllib.request.urlopen(req, timeout=30) as r:
        return json.loads(r.read().decode())

try:
    train = post_json('http://127.0.0.1:5001/api/train', {})
    print('train:', train)
    pred = post_json('http://127.0.0.1:5001/api/predict', {
        'price': 25, 'promotion': 1, 'is_weekend': 1, 'temp': 22,
        'quality': 8, 'competitors': 2, 'holiday': 0, 'location': 7,
        'weather_score': 1.0,
    })
    print('predict:', pred)
except Exception as e:
    print('请先终端启动: python mini_predict_app.py')
    print(type(e).__name__, e)
````
