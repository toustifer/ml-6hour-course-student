## 1. HTML 很粗的结构

```html
<button onclick="startTrain()">开始训练</button>
<div id="trainResult"></div>
```

- `button`：按钮
- `onclick`：点击时调用的函数名
- `div`：一块显示区域

## 2. JavaScript 调用 API（完整版里的思路）

```javascript
async function startTrain() {
  const r = await fetch('/api/train', { method: 'POST' });
  const j = await r.json();
  // j.r2, j.importance ...
}
```

- `fetch`：浏览器向服务器要数据
- `await`：等结果返回
- `j.r2`：用 JSON 里的字段

迷你版 `mini_predict_app.py` 底部 `<script>` 也是同一模式，行数更少，**建议先读迷你版脚本**。
