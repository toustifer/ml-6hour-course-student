## 1.2 相对路径和绝对路径

在读取文件之前，先搞清楚一个很重要的概念：**路径**。

### 绝对路径
绝对路径就是“完整地址”，从电脑里的大位置一路写到文件名。

例如：

```python
D:/obi/熵学院/授课/机器学习-6小时/教学课程/day1/day1_cafe_sales.csv
```


### 相对路径
相对路径不是从整台电脑最外面开始写，而是从**当前工作位置**开始写。

例如，如果我们现在打开的就是 `day1` 文件夹，那么：

```python
"../../day1_cafe_sales.csv"
```

就表示：**在当前文件夹里找这个文件**。

---

上课时更常用相对路径，因为它更短，也更适合大家用同一套课程文件。

## 1.3 相对路径是从哪里开始算的

这件事特别容易卡住。

请先记住一句话：

> 相对路径不是从 notebook 文件名开始算，
> 而是从**当前打开的工作文件夹**开始算。

这次课程里，我们要求大家都用 VS Code 打开 `教学课程/day1` 这个文件夹。  
所以当我们写：

```python
day1_cafe_sales.csv
```

意思就是：

- 先看当前打开的位置是不是 `day1`
- 如果是，就在这个文件夹里找 `day1_cafe_sales.csv`

如果你打开错了位置，比如打开到了更上层文件夹，电脑就可能找不到这个文件。

所以“找不到 csv”时，先别急着改代码，先检查：

1. 左边文件栏里是不是打开了 `day1` 文件夹
2. `day1_cafe_sales.csv` 是不是和 notebook 在同一个教学目录里

这比乱改文件名更重要。

---

正因为不同同学有时会打开错文件夹，下面的代码才会准备几个可能的位置，尽量自动帮我们找到 csv 文件。



````python
from pathlib import Path
import pandas as pd

csv_path = Path("day1_cafe_sales.csv")

candidate_paths = [
    Path('day1_cafe_sales.csv'),
    Path('day1') / 'day1_cafe_sales.csv',
    Path('教学课程') / 'day1' / 'day1_cafe_sales.csv',
]

for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day1_cafe_sales.csv')

print(csv_path.resolve())
````

## 2. 用 pandas 读取数据

今天最重要的一行代码之一：`pd.read_csv(...)`。

````python
df = pd.read_csv(csv_path)
print(type(df))
````
