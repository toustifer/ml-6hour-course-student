## 3. 什么是数据表

把数据表想象成一张很大的 Excel：

- 每一行：代表一天
- 每一列：代表一个信息项
- 每一个单元格：代表某一天的某个具体数值

例如：

- `weather_label`：这一天的天气
- `price`：这一天的商品价格
- `sales`：这一天的总营收

## 5. 行和列怎么记

- 第 1 行：第 1 天的经营情况
- `day` 列：第几天
- `weather_label` 列：天气文字
- `price` 列：价格
- `sales` 列：当天营收

一句话：

> 行看“某一天”，列看“某一类信息”。

## 6. 业务词汇翻译

把字段名翻译成日常语言：

| 字段名 | 日常含义 |
| --- | --- |
| `temp` | 当天温度 |
| `promotion` | 有没有做促销 |
| `is_weekend` | 是不是周末 |
