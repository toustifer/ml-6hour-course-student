## 1. 先搞清楚这些工具分别干什么

| 工具 | 一句话理解 | 上课时主要用来做什么 |
| --- | --- | --- |
| Python | 会“算”的语言 | 运行我们写的代码 |
| Jupyter Notebook | 一格一格写代码的本子 | 上课主界面：写、运行、看输出 |

它们不是互相替代，而是一起用。

## 4. 路线 B：安装 conda（Miniconda / Anaconda）

### 4.1 conda 是什么？

**conda** 是一个**环境管理 + 包管理**工具。

一句话：

- 可以装 Python
- 可以给不同课程建“独立房间”（环境），互不打架
- 可以在房间里安装 `pandas`、`matplotlib` 等包

为什么课堂常用它？

- 避免“我电脑上有三个 Python，装包装错地方”
- 老师可以说：“大家一起用 `day1_ml` 这个环境”
- 以后 Day 2～5 再加包，也装进同一个环境即可

### 4.2 Miniconda 还是 Anaconda？

| 产品 | 特点 | 课堂建议 |
| --- | --- | --- |
| **Miniconda** | 体积小，只带 conda + 基础 Python，包自己装 | **更推荐**（快、干净） |
| **Anaconda** | 体积大，自带很多数据科学包 | 机房网速慢、想少敲命令时可用 |

两者都提供 `conda` 命令，**用法一样**。本课程命令以 conda 为准。

### 4.3 下载地址（具体直链）

#### 先确认自己的芯片

点左上角苹果菜单 →「关于本机」：

- 写着 **Apple M1 / M2 / M3 / M4** → 选 **arm64 / Apple Silicon**
- 写着 **Intel** → 选 **x86_64**

#### Miniconda（推荐，双击 `.pkg` 安装）

| 芯片 | 安装包直链 |
| --- | --- |
| **Apple 芯片（M 系列）** | https://repo.anaconda.com/miniconda/Miniconda3-latest-MacOSX-arm64.pkg |
| **Intel 芯片** | https://repo.anaconda.com/miniconda/Miniconda3-latest-MacOSX-x86_64.pkg |

说明页 / 其他版本：https://docs.conda.io/en/latest/miniconda.html  
完整文件列表：https://repo.anaconda.com/miniconda/

> 若学校网络打不开 `repo.anaconda.com`，可改用页面上的说明安装，或请老师提前拷贝 `.pkg` 到 U 盘。

#### Anaconda（体积大，可选）

- 官网下载页：https://www.anaconda.com/download
- 选 **macOS**，再选 **Apple Silicon** 或 **Intel**

### 4.4 Mac 安装注意

1. 下载 `.pkg` 后双击安装，路径建议用默认
2. 装完后**新开一个终端窗口**（很重要，否则可能找不到 `conda`）
3. 若第一次打开终端提示是否初始化 shell，选 **yes**
4. 若输入 `conda` 仍提示找不到，可手动初始化一次（zsh 为 macOS 默认）：

```text
# Miniconda 常见路径（按你实际安装位置调整）
~/miniconda3/bin/conda init zsh
# 或 Anaconda：
# ~/anaconda3/bin/conda init zsh
```

然后**关掉终端再打开**，提示符前可能会出现 `(base)`。

### 4.5 确认 conda 装好了

在终端中输入：

```text
conda --version
python --version
```

看到版本号即可，例如：

```text
conda 24.x.x
Python 3.12.x
```


## 5. 用 conda 创建课程环境（核心步骤）

装完 conda 后，**不要直接在 base 里乱装一堆包**（可以，但不推荐）。

更好的习惯：为这门课建一个独立环境，名字叫 `day1_ml`。

### 5.1 创建环境并装上课包

在**终端**中**逐行**输入：

```text
conda create -n day1_ml python=3.11 -y
conda activate day1_ml
conda install pandas matplotlib jupyter -y
```

含义：

| 命令 | 意思 |
| --- | --- |
| `conda create -n day1_ml python=3.11` | 新建名为 `day1_ml` 的环境，Python 版本 3.11 |
| `conda activate day1_ml` | 进入这个环境（提示符前会出现 `(day1_ml)`） |
| `conda install pandas matplotlib jupyter -y` | 在环境里安装上课要用的包 |

如果 `conda install` 很慢，可以先配置清华镜像（可选）：

```text
conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/main
conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free
conda config --set show_channel_urls yes
```

然后再执行 `conda install ...`。

### 5.2 确认环境里的包

环境激活后（提示符有 `(day1_ml)`）：

```text
conda list pandas
python -c "import pandas, matplotlib; print('ok')"
```

输出 `ok` 就说明环境可用。

### 5.3 每次上课怎么进入环境？

1. 打开 **终端（Terminal）**
2. 输入：

```text
conda activate day1_ml
```

3. 再打开 VS Code，或在 VS Code 里选择 `day1_ml` 对应的 Python 内核

### 5.4 conda 和 pip 的关系（先记住这三句）

1. **优先用 conda 装**：`conda install 包名`
2. conda 没有的包，再在**已 activate 的环境里**用：`pip install 包名`
3. 不要在没 activate 的情况下到处 `pip install`，否则包会装错 Python

本课 Day 1 只需要：`pandas`、`matplotlib`、`jupyter`，用 conda 就够。

## 11. 常见报错与处理（Mac）

### 11.1 `python3` / `conda`：command not found

- 原因：没装好，或终端还没加载 PATH，或 conda 没 `init`
- 处理：
  - 路线 A：重新安装官网 Python，装完**关掉终端再开**，用 `python3` 而不是 `python`
  - 路线 B：新开终端；仍不行则执行 `~/miniconda3/bin/conda init zsh`（路径按实际安装调整），再重启终端

### 11.2 `No module named 'pandas'`

- 原因：包还没装，或装到了另一个 Python / 另一个 conda 环境里
- 处理：
  - 路线 A：`python3 -m pip install pandas matplotlib`
  - 路线 B：先 `conda activate day1_ml`，再 `conda install pandas matplotlib -y`
  - 然后在 notebook 右上角**重新选择**同一个环境的内核

### 11.3 notebook 选不到 `day1_ml`

- 先确认环境存在：`conda env list` 里应有 `day1_ml`
- VS Code 已装 Python + Jupyter 扩展
- `Command + Shift + P` → `Python: Select Interpreter` → 选 `day1_ml`
- 仍没有：重启 VS Code 再试

### 11.4 `conda activate` 失败

- 先确认已执行过 `conda init zsh`（或 `conda init bash`），并**重启终端**
- 若提示 shell 未初始化：按终端里 conda 给出的提示操作
- 也可以在 VS Code 终端里直接选带 `day1_ml` 的解释器，少用手动 activate

### 11.5 运行格子没有反应 / 一直转圈

- 先看右上角是否已经选中内核
- 点 `重启` 后再运行一次
- 确认装的是 Python 扩展 + Jupyter 扩展

### 11.6 找不到 `day1_cafe_sales.csv`

- 原因：VS Code 打开的不是 `day1` 文件夹
- 处理：重新 `打开...`，选到包含 CSV 和 notebook 的那个目录

### 11.7 画图中文乱码

- Day 1 后面模块会设置中文字体
- 先保证代码能跑；字体问题后面再统一处理
- Mac 上后面可用系统字体如 `PingFang SC` / `Heiti SC`（后续模块会写）

- 看 `sys.executable` 打印的路径，确认自己跑的是哪一份 Python

### 11.9 安装包时提示「无法打开 / 身份不明的开发者」

- **系统设置** → **隐私与安全性** → 「仍要打开」

### 11.10 Apple 芯片装错了 Intel 包

- 症状：conda / 某些包异常慢或装不上
- 处理：卸载后改装 **arm64 / Apple Silicon** 版 Miniconda 或 Anaconda，再建 `day1_ml`
