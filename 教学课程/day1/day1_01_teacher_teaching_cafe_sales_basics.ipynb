## 3. 读取 CSV

前面已经有了变量、列表、if、for。

现在第一次把真实数据表读进 Python。

![CSV 到 DataFrame 示意图](day1_csv_to_dataframe_demo.png)

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

candidate_paths = [
    Path('day1_cafe_sales.csv'),
    Path('day1') / 'day1_cafe_sales.csv',
    Path('教学课程') / 'day1' / 'day1_cafe_sales.csv',
]

for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day1_cafe_sales.csv，请确认它和 notebook 位于教学课程/day1 目录。')

df = pd.read_csv(csv_path)
print('数据文件位置:', csv_path.resolve())
print('数据表形状:', df.shape)
df.head()
```

## 4. 快速认识数据

![看懂数据表的四个常用工具](day1_table_tools_overview_demo.png)

![数值列与文字列对比图](day1_numeric_vs_text_columns_demo.png)

```python
print('数据形状:', df.shape)
print('列名列表:', df.columns.tolist())
df.info()
```

```python
df.describe().round(2)
```

## 5. 最基础业务统计

```python
summary = {
    '总天数': len(df),
    '平均营收': round(df['sales'].mean(), 2),
    '最高营收': int(df['sales'].max()),
    '最低营收': int(df['sales'].min()),
    '平均售价': round(df['price'].mean(), 2),
}

for key, value in summary.items():
    print(f'{key}: {value}')
```

## 6. 用 groupby 看天气和营收

![groupby 分组统计示意图](day1_groupby_explained_demo.png)

问题：不同天气下，平均营收一样吗？

```python
weather_order = ['晴', '多云', '阴', '小雨', '大雨']
avg_sales_by_weather = df.groupby('weather_label')['sales'].mean().reindex(weather_order).round(2)
avg_sales_by_weather
```

## 7. 最小可视化

![折线图 vs 柱状图](day1_line_vs_bar_chart_demo.png)

- 折线图：看趋势
- 柱状图：看分组对比

```python
plt.figure(figsize=(12, 4))
plt.plot(df['day'], df['sales'], color='#6f4e37', linewidth=1.8)
plt.title('每日营收趋势')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
avg_sales_by_weather.plot(kind='bar', color='#c89b74')
plt.title('不同天气下的平均营收')
plt.xlabel('天气')
plt.ylabel('平均营收')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```
