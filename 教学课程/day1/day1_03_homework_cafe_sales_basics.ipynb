## 任务 1：Python 基础

1. 定义变量 `shop_name` 和 `today_sales` 并打印
2. 创建列表 `sales_list = [2500, 2800, 3300]`
3. 用 for + if：营收 >= 3000 打印“较高”，否则“一般”

```python
# 在这里写任务 1
shop_name = '熵学院咖啡馆'
today_sales = 3000
print(shop_name, today_sales)

sales_list = [2500, 2800, 3300]
for s in sales_list:
    if s >= 3000:
        print(s, '较高')
    else:
        print(s, '一般')
```

## 任务 2：读取数据并查看前几行

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

candidate_paths = [
    Path('day1_cafe_sales.csv'),
    Path('day1') / 'day1_cafe_sales.csv',
    Path('教学课程') / 'day1' / 'day1_cafe_sales.csv',
]

for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day1_cafe_sales.csv，请确认它和 notebook 位于教学课程/day1 目录。')

df = pd.read_csv(csv_path)
print('数据文件位置:', csv_path.resolve())
print('数据表形状:', df.shape)
df.head(10)
```

## 任务 3：认识数据

输出：

- 形状
- 列名
- `info()`
- `describe()`

```python
print(df.shape)
print(df.columns.tolist())
df.info()
df.describe().round(2)
```

## 任务 4：基础统计

计算总天数、平均营收、最高营收、最低营收。

```python
print('总天数:', len(df))
print('平均营收:', round(df['sales'].mean(), 2))
print('最高营收:', int(df['sales'].max()))
print('最低营收:', int(df['sales'].min()))
```

## 任务 5：groupby

计算不同天气的平均营收。

![groupby 分组统计示意图](day1_groupby_explained_demo.png)

```python
weather_order = ['晴', '多云', '阴', '小雨', '大雨']
avg_sales_by_weather = df.groupby('weather_label')['sales'].mean().reindex(weather_order).round(2)
avg_sales_by_weather
```

## 任务 6：画两张图

![折线图 vs 柱状图](day1_line_vs_bar_chart_demo.png)

1. 每日营收折线图
2. 天气平均营收柱状图

```python
plt.figure(figsize=(12, 4))
plt.plot(df['day'], df['sales'], color='#6f4e37', linewidth=1.8)
plt.title('每日营收趋势')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
avg_sales_by_weather.plot(kind='bar', color='#c89b74')
plt.title('不同天气下的平均营收')
plt.xlabel('天气')
plt.ylabel('平均营收')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```

## 任务 7：写一句结论

用一句话回答：哪种天气平均营收更高？你从图上看到了什么？

```python
best_weather = avg_sales_by_weather.idxmax()
print('平均营收最高的天气是:', best_weather)
print('我的结论：')  # 在下一行用 print 写出你的结论
```
