## 任务 1

读取 `day1_cafe_sales.csv`，并打印前 5 行。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

candidate_paths = [Path('day1_cafe_sales.csv'), Path('day1') / 'day1_cafe_sales.csv', Path('教学课程') / 'day1' / 'day1_cafe_sales.csv']
for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day1_cafe_sales.csv')

df = pd.read_csv(csv_path)
df.head()
```

## 任务 2

查看：

```python
print(df.shape)
print(df.columns.tolist())
df.info()
```

## 任务 3

计算：

- 总天数
- 平均营收
- 最高营收
- 最低营收

```python
print('总天数:', len(df))
print('平均营收:', round(df['sales'].mean(), 2))
print('最高营收:', int(df['sales'].max()))
print('最低营收:', int(df['sales'].min()))
```

## 任务 4

按天气分组，比较不同天气下的平均营收。

```python
weather_order = ['晴', '多云', '阴', '小雨', '大雨']
avg_sales_by_weather = df.groupby('weather_label')['sales'].mean().reindex(weather_order).round(2)
avg_sales_by_weather
```

## 任务 5 参考图：什么时候用折线图 / 柱状图

画图前先确认选择：

![折线图和柱状图选择示意图](day1_line_vs_bar_chart_demo.png)

- 看变化过程：折线图
- 看类别比较：柱状图

```python
plt.figure(figsize=(12, 4))
plt.plot(df['day'], df['sales'], color='#6f4e37', linewidth=1.8)
plt.title('每日营收趋势')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

## 任务 6

画一张不同天气平均营收柱状图。

```python
plt.figure(figsize=(8, 4))
avg_sales_by_weather.plot(kind='bar', color='#c89b74')
plt.title('不同天气下的平均营收')
plt.xlabel('天气')
plt.ylabel('平均营收')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```
