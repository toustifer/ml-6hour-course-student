## 4. 数字：可以计算

常见数字有两种：

- 整数：`3200`
- 小数：`28.5`

数字可以直接做加减乘除。

````python
sales_a = 3000
sales_b = 3500

print('两天营收总和:', sales_a + sales_b)
print('两天营收差值:', sales_b - sales_a)
print('两天营收平均:', (sales_a + sales_b) / 2)
````

## 5. 字符串：文字信息

字符串就是文字，要用引号包起来。

单引号和双引号都可以：

- `'晴'`
- `"多云"`

````python
weather_text = '多云'
shop_name = '熵学院咖啡馆'

print('天气文字:', weather_text)
print('店名:', shop_name)
print('拼接结果:', shop_name + ' 今天天气是 ' + weather_text)
````

### 重点提醒

- 数字可以算：`3000 + 200`
- 文字通常用来表示名称、类别：`'晴'`
- 不要把文字和数字混着直接相加，例如 `'晴' + 3000` 会报错

````python
# 正确：数字和数字
print(3000 + 200)

# 正确：文字和文字
print('今天天气：' + '晴')

# 如果要把数字和文字拼在一起，先把数字变成文字
print('今天营收：' + str(3200))
````

## 6. 布尔值：对 / 错

布尔值只有两个结果：

- `True`：真
- `False`：假

后面做判断时会经常用到。

````python
sales = 3600

print(sales >= 3000)
print(sales < 2000)
print(sales == 3600)
````

常见比较符号：

- `>` 大于
- `<` 小于
- `>=` 大于等于
- `<=` 小于等于
- `==` 等于
- `!=` 不等于

## 7. 列表：按顺序放多个值

列表用方括号 `[]`。

例如一周的营收，就可以放进一个列表。

````python
week_sales = [2800, 'abc', 2950, 3300, 3600]

print('列表内容:', week_sales)
print('第 1 个值:', week_sales[0])
print('第 2 个值:', week_sales[1])
print('最后一个值:', week_sales[-1])
print('一共有几个值:', len(week_sales))
print('最后一个值:', week_sales[len(week_sales) - 1])
````

## 8. 字典：名字对应含义

字典用花括号 `{}`。

它适合表示“名字 -> 含义”。

例如天气编码：

- `0` 表示晴
- `1` 表示多云

````python
weather_map = {
    0: '晴',
    1: '多云',
    '123': '阴',
    3: '小雨',
    4: '大雨'
}

print(weather_map)
print('编码 0 对应:', weather_map[0])
print('编码 3 对应:', weather_map[3])
print("编码 '123' 对应:", weather_map['123'])
````

## 9. if 判断：条件成立就做什么

if 的意思是：

- 如果条件成立，就执行这一段
- 否则，就执行另一段

````python
sales = 2900

if sales >= 3000:
    print('今天营收比较高')
else:
    a = 10
    print('今天营收一般')
    print("上面:", a)
print("下面:",a)
````

### 再看一个例子

````python
weather = '小雨'

if weather == '晴':
    print('适合到店消费')
elif weather == '多云':
    print('天气还不错')
else:
    print('可能会影响客流')
````

## 10. for 循环：重复做一件事

循环就是把类似的动作重复做很多次。

最常见写法：把列表里的每个值依次拿出来处理。

````python
sales_list = [2600, 2800, 3000]



for item in sales_list:
    print('一条营收记录:', item)
````

### 循环 + 判断

````python
sales_list = [2400, 3100, 2900, 3600]

for item in sales_list:
    print('一条营收记录:', item)
    if item >= 3000:
        print(item, '-> 较高')
    else:
        print(item, '-> 一般')
````

## 11. 简单函数：把常用逻辑打包起来

前面我们已经写过很多次“判断营收高不高”的逻辑。

如果一段逻辑以后还要反复使用，就可以把它打包成一个小工具，这就叫**函数**。

你现在先把函数理解成：

> 给一段常用代码取名字，以后直接调用。

今天只要求会看、会改、会调用最简单的函数。

````python
def sales_level(value):
    if value >= 3000:
        return '较高'
    return '一般'

print(sales_level(3200))
print(sales_level(2400))
print(sales_level(3000))
````
