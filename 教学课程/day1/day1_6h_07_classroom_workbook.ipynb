```python
# 1) 变量
student_name = '请改成你的名字'
today_sales = 3000
print(student_name)
print(today_sales)

# 2) 列表
sample_sales = [2500, 2800, 3100, 3300]
print(sample_sales)
print('第 1 个值:', sample_sales[0])
print('长度:', len(sample_sales))

# 3) 字典
weather_map = {0: '晴', 1: '多云', 3: '小雨'}
print(weather_map)
print('0 对应:', weather_map[0])

# 4) if 判断
if today_sales >= 3000:
    print('营收较高')
else:
    print('营收一般')

# 5) for 循环
for s in sample_sales:
    print('营收记录:', s)
```

### 练习 2 加分题

请写一个函数 `is_high_sales(value)`：

- 如果 `value >= 3000`，返回 `'较高'`
- 否则返回 `'一般'`

并测试 `2800` 和 `3500`。

```python
def is_high_sales(value):
    if value >= 3000:
        return '较高'
    return '一般'

print(is_high_sales(2800))
print(is_high_sales(3500))
```

## 练习 3：读取 CSV

**对应模块：模块 3**

只有学完 Python 基础和“读取 CSV”之后，再做这一题。

请把 day1 数据读进来，并显示前 5 行。

```python
from pathlib import Path
import pandas as pd

candidate_paths = [Path('day1_cafe_sales.csv'), Path('day1') / 'day1_cafe_sales.csv', Path('教学课程') / 'day1' / 'day1_cafe_sales.csv']
for path in candidate_paths:
    if path.exists():
        csv_path = path
        break
else:
    raise FileNotFoundError('未找到 day1_cafe_sales.csv')

df = pd.read_csv(csv_path)
df.head()
```

## 练习 4：基础统计

**对应模块：模块 5**

请计算：

- 总天数

- 最高营收

```python
print('总天数:', len(df))
print('平均营收:', round(df['sales'].mean(), 2))
print('最高营收:', int(df['sales'].max()))
```

## 练习 5：分组分析

**对应模块：模块 5**

请比较不同天气下的平均营收。

### 练习 5 参考图：groupby 在做什么

![groupby 分组统计示意图](day1_groupby_explained_demo.png)

- 先按天气分组
- 再对每组求平均营收

```python
weather_order = ['晴', '多云', '阴', '小雨', '大雨']
df.groupby('weather_label')['sales'].mean().reindex(weather_order).round(2)
```
